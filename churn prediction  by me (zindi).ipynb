```python
continuous = ['MONTANT','FREQUENCE_RECH', 'REVENUE','ARPU_SEGMENT','FREQUENCE','DATA_VOLUME','ON_NET', 'ORANGE', 'TIGO', 'FREQ_TOP_PACK']
def fillna_with_mode():
    for i in continuous:
        x = combined[i].mode()[0]
        combined[i] = combined[i].fillna(x)
    return combined
    
def fillna_with_mean():
    for i in continuous:
        x = combined[i].mean()
        combined[i] = combined[i].fillna(x)
    return combined
        
```
